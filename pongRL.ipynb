```python
#CLASSES

class Perceptron:
    def __init__(self, weights:np.ndarray, biases:np.ndarray, activation:object):
        self.weights = weights
        self.biases = biases
        self.activation = activation

    def __call__(self, inputs:list):
        return self.activation(sum(inputs * self.weights + self.biases))
    
    def __str__(self):
        return f'{self.activation.__name__}(w{self.weights} + b{self.biases})'
    
class Network:
    def __init__(self):
        self.layers = []
        self.output=None

    def __call__(self, inputs:list):
        for layer in self.layers:
            inputs = [perc(inputs) for perc in layer]
        return inputs

    def __str__(self):
        res=''
        for layer in self.layers:
            res+='['
            for node in layer:
                res+=str(node)+(' ' if node != layer[-1] else '')
            res+=']\n'
        return res

class DNA:pass
```

```python
#FUNCTIONS

def complete_bin(binary:str, digits:int, isneg:bool|None=None, unsigned:bool=False)->str:
    binary = '0'*(digits-len(binary)-(1 if not isneg == None else 0)) + binary
    return (('1' if isneg else '0') if not isneg==None else '') + binary

def n2bin(n:int)->str:
    return bin(n).split('b')[1]

twobit2dna = {
    '00':'A',
    '01':'C',
    '10':'G',
    '11':'T'
}

dna2bits = {
    'A':'00',
    'C':'01',
    'G':'10',
    'T':'11'
}

def bin2dna(binary:str)->str:
    res=''
    for i in range(0, len(binary), 2):
        word = binary[i:i+2]
        res+=twobit2dna[word]
    return res

def float2dna(num:int|float)->str:
    num = round(float(num), 2)
    
    integer = int(str(num).split('.')[0])
    try:
        first, second = str(num).split('.')[1]
    except ValueError:
        first, second = str(num).split('.')[1], '0'
    first, second = int(first), int(second)

    integer_bin = n2bin(integer)
    first_bin = n2bin(first)
    second_bin = n2bin(second)

    if len(integer_bin) > 7: raise Exception('too big weight value')

    integer_complete = complete_bin(integer_bin, 8, num<0)
    first_complete = complete_bin(first_bin, 4)
    second_complete = complete_bin(second_bin, 4)

    return bin2dna(integer_complete) + bin2dna(first_complete) + bin2dna(second_complete)

def dna2float(dna:str)->float:
    integer_dna = dna[:4]
    first_dna = dna[4:6]
    second_dna = dna[6:]

    integer_bin=''.join([dna2bits[some] for some in integer_dna])
    first_bin=''.join([dna2bits[some] for some in first_dna])
    second_bin=''.join([dna2bits[some] for some in second_dna])

    integer = int(integer_bin[1:], 2) * (-1 if int(integer_bin[0]) else 1)
    first = int(first_bin, 2)
    second = int(second_bin, 2)

    return float(f'{integer}.{first}{second}')

def perceptron2dna(perceptron:Perceptron)->str:pass

def network2dna(network:Network)->str:pass

def dna2perceptron(dna:str)->Perceptron:pass

def dna2network(dna:str)->Network:pass
```
